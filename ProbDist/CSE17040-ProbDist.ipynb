###### Problem

Imagine that we are data scientists tasked with improving the ROI (Return on Investment) of our company’s call center, where employees attempt to cold call potential customers and get them to purchase our product. You look at some historical data and find the following:
    <ul>
         <li>The typical call center employee completes on average 50 calls per day.</li>
         <li>The probability of a conversion (purchase) for each call is 4%.</li>
         <li>The average revenue to your company for each conversion is ₹20.</li>
         <li>The call center you are analyzing has 100 employees.</li>
         <li>Each employee is paid ₹200 per day of work.</li>
    </ul>

```python
import numpy as np
import matplotlib.pyplot as p
import seaborn as sns
from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
```

```python
a = 1
b = 50
employees = 100
wage = 200
n = 50
p = 0.04
revenue = 100
sims = 10000
mu = 2
```

## Original Simulation

### Uniform Distribution

```python
uniform_conversions = [np.sum(uniform.rvs(size=employees, loc = a, scale=b)) for _ in range(sims)]
uniform_profits = np.array(uniform_conversions)*revenue - employees*wage
sns.distplot(uniform_profits)
```

```python
print('Average Conversions per Employee               : ' + str(round(np.mean(uniform_conversions), 2)))
print('Standard Deviation of Conversions per Employee : ' + str(round(np.std(uniform_conversions))))
print('Total Conversions                              : ' + str(np.sum(uniform_conversions)))
print('Total Revenues                                 : ' + str(np.sum(uniform_conversions)*revenue))
print('Total Expense                                  : ' + str(employees*wage))
print('Total Profits                                  : ' + str(np.sum(uniform_conversions)*revenue - employees*wage))
```

### Normal Distribution

```python
normal_conversions = [np.sum(uniform.rvs(size=employees, loc = a, scale=b)) for _ in range(sims)]
normal_profits = np.array(normal_conversions)*revenue - employees*wage
sns.distplot(normal_profits)
```

```python
print('Average Conversions per Employee               : ' + str(round(np.mean(normal_conversions), 2)))
print('Standard Deviation of Conversions per Employee : ' + str(round(np.std(normal_conversions))))
print('Total Conversions                              : ' + str(np.sum(normal_conversions)))
print('Total Revenues                                 : ' + str(np.sum(normal_conversions)*revenue))
print('Total Expense                                  : ' + str(employees*wage))
print('Total Profits                                  : ' + str(np.sum(normal_conversions)*revenue - employees*wage))
```

### Bernoulli Distribution

```python
bern_conversions = [np.sum(bernoulli.rvs(size=employees, p=p)) for _ in range(sims)]
bern_profits = np.array(bern_conversions)*revenue - employees*wage
sns.distplot(bern_profits)
```

```python
print('Average Conversions per Employee               : ' + str(round(np.mean(bern_conversions), 2)))
print('Standard Deviation of Conversions per Employee : ' + str(round(np.std(bern_conversions))))
print('Total Conversions                              : ' + str(np.sum(bern_conversions)))
print('Total Revenues                                 : ' + str(np.sum(bern_conversions)*revenue))
print('Total Expense                                  : ' + str(employees*wage))
print('Total Profits                                  : ' + str(np.sum(bern_conversions)*revenue - employees*wage))
```

### Binomial Distribution

```python
binom_conversions = [np.sum(binom.rvs(size=employees, p=p, n=n)) for _ in range(sims)]
binom_profits = np.array(binom_conversions)*revenue - employees*wage
sns.distplot(binom_profits)
```

```python
print('Average Conversions per Employee               : ' + str(round(np.mean(binom_conversions), 2)))
print('Standard Deviation of Conversions per Employee : ' + str(round(np.std(binom_conversions))))
print('Total Conversions                              : ' + str(np.sum(binom_conversions)))
print('Total Revenues                                 : ' + str(np.sum(binom_conversions)*revenue))
print('Total Expense                                  : ' + str(employees*wage))
print('Total Profits                                  : ' + str(np.sum(binom_conversions)*revenue - employees*wage))
```

### Poisson Distribution

```python
poisson_conversions = [np.sum(poisson.rvs(mu=mu, size=employees)) for _ in range(sims)]
poisson_profits = np.array(poisson_conversions)*revenue - employees*wage
sns.distplot(poisson_profits)
```
