```python
from fcmeans import FCM
import pandas as pd
import matplotlib.pyplot as plt
from seaborn import scatterplot as scatter
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('SPECTF_New.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
columns = list(df.columns)
features = columns[:len(columns)-1]
class_labels = list(df[columns[-1]])
dfMod = df[features]
```

```python
df['Class'].unique()
```

```python
pca = PCA(n_components=2)

pComp = pca.fit_transform(dfMod)
principalDf = pd.DataFrame(data = pComp
             , columns = ['PC1', 'PC2'])
```

```python
principalDf.head()
```

```python
plt.figure(figsize=(10,10))                                              
plt.scatter(list(principalDf.iloc[:,0]), list(principalDf.iloc[:,1]), marker='o')       
plt.axis('equal')                                                                 
plt.xlabel('PC1', fontsize=16)                                                 
plt.ylabel('PC2', fontsize=16)                                                 
plt.title('Plot', fontsize=22)                                            
plt.grid()                                                                         
plt.show()
```

```python
fcm = FCM(n_clusters=2)
fcm.fit(principalDf)
```

```python
fcm_centers = fcm.centers
fcm_labels = fcm.u.argmax(axis=1)
```

```python

scatter(pComp[:,0], pComp[:,1])

plt.show()
```

```python
scatter(pComp[:,0], pComp[:,1], hue=fcm_labels)
scatter(fcm_centers.iloc[:,0], fcm_centers.iloc[:,1],marker="s",s=200)
plt.show()
```
